```python
### 3D connectivity metrics grphics ###

rdir='/media/chris/kinoto/indicators/connectivity_paper/2D_1024_3D_128/3D/'
folder = ['3D_ind_lc1.5', '3D_ind_lc2', '3D_ind_lc3', '3D_ind_lc4'] #, '3D_ind_3', '3D_ind_4']

L=128
n=20

lc = [1.5,2,3,4]

x, y =np.zeros((n)),np.zeros((n))
xind='PPHA'
markers = ['x', 'o']
color = ['r', 'g', 'b', 'darkblue', 'black', 'grey', 'gray', 'slategrey', 'darkred', 'yellow']

yinds=['VOLALE','GAMMA','Plen','S','P', 'spanning', 'npx', 'npy', 'npz']

for yind in yinds:
    for ilc in range(len(folder)):
        for i in range(n):
            res=np.load(rdir + folder[ilc] + '/' + str(i) + '/ConnectivityMetrics/'+str(L)+'.npy',allow_pickle=True).item()
            np.squeeze(res)
            x[i]=np.squeeze(res[xind])
            y[i]=np.squeeze(res[yind])
        plt.plot(x,y,label="lc=" + str(lc[ilc]))
        plt.legend()
        plt.xlabel(xind)
        plt.ylabel(yind)
        plt.title('2D L=128')
    plt.show()
    plt.close()
```
