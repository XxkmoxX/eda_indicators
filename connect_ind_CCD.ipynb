```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

rdir = '/media/chris/kinoto/indicators/ccd/'
ddir = ['2D/','3D/']
cdir = ['ics/', 'hcs/', 'lcs/']

d = [2, 3]
p = np.linspace(0.4, 0.8, 20)
ec = [1, 2, 3]
ec_label = ['ICS', 'HCS', 'LCS']
lc = [[1.5, 2, 3], [2, 3, 4]]
L = [1024, 128]

inds = ['PPHA', 'VOLALE', 'ZNCC', 'GAMMA', 'spanning', 'npz', 'npy', 'npx', 'Plen', 'S', 'P', 'PlenX', 'SX', 'PX']
```

```python
### 2D ###

yinds = inds[1:len(inds)+1]
x1 = inds[0]
x2 = np.linspace(0.4, 0.8, 20)

x, y = np.zeros(len(p)), np.zeros(len(p))

color = ['green', 'red', 'blue']
marker = ['x', 'o']

for yind in yinds:
    for iec in range(len(ec)):
        for ip in range(len(p)):
            res = np.load(rdir + ddir[0] + cdir[iec] + str(ip * 10) + '/ConnectivityMetrics/' + '1024.npy', allow_pickle=True).item()
            x[ip] = np.squeeze(res[x1])
            y[ip] = np.squeeze(res[yind])
        print(x, y)
        plt.plot(x, y, label = ec_label[iec], color = color[iec], marker = marker[0])
        plt.legend()
        plt.xlabel(x1)
        plt.ylabel(yind)
        plt.title('2D, L=1024, lc=2')
    plt.show()
```

```python
### 3D ###
yinds = inds[1:len(inds)+1]
x1 = inds[0]
x2 = np.linspace(0.4, 0.8, 20)
x, y = np.zeros(len(p)), np.zeros(len(p))

color = ['darkgreen', 'darkred', 'darkblue']
marker = ['x', 'o']

for yind in yinds:
    for iec in range(len(ec)):
        for ip in range(len(p)):
            res = np.load(rdir + ddir[1] + cdir[iec] + str(ip * 10) + '/ConnectivityMetrics/' + '128.npy', allow_pickle=True).item()
            x[ip] = np.squeeze(res[x1])
            y[ip] = np.squeeze(res[yind])
        plt.plot(x, y, label = ec_label[iec], color = color[iec], marker = marker[0])
        plt.legend()
        plt.xlabel(x1)
        plt.ylabel(yind)
        plt.title('3D, L=128' + ', ' + 'lc=1.5')
    plt.show()
```
